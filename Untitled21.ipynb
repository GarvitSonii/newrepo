# import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

# defination of gaussian function

```python
def gaussian_2d(x, y, x0, y0, sigma_x, sigma_y, A=1):
    return A * np.exp(-(((x - x0)**2) / (2 * sigma_x**2) + ((y - y0)**2) / (2 * sigma_y**2)))
```

# scale function

```python
def scale(image):
    current_max = np.max(image)
    current_min = np.min(image)
    scaled_image = ((image-current_min) / (current_max-current_min)) * 255
    return scaled_image
    
```

# defination fourier transform function

```python
def fourier_transform(image):
    shifted_image = image * np.fromfunction(lambda x, y: (-1) ** (x + y), image.shape)
    fft_image = np.fft.fft2(shifted_image)
    return fft_image
```

# defination inverse fourier transform function

```python
def inverse_fourier_transform(spectrum):
    ifft_image = np.fft.ifft2(spectrum)
    ifft_image = ifft_image * np.fromfunction(lambda x, y: (-1) ** (x + y), ifft_image.shape)
    return ifft_image
```

# defination of image

```python
def npimage(s):
    image_ = Image.open(s)
    image_c = image_.convert('L')
    image = np.array(image_c)
    return image
```

# defination of ideal low pass filter

```python
def ideal_lpf(image, radius):
    filter_ = np.zeros(image.shape)
    center = tuple(map(lambda x: x // 2, image.shape)) 
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            if (i - center[0]) ** 2 + (j - center[1]) ** 2 <= radius ** 2:
                filter_[i, j] = 1 
                
    ft_image = fourier_transform(image)
    ft_image = ft_image * filter_
    ift_image = scale(inverse_fourier_transform(ft_image))
    
    return ift_image
```

# defination of gaussian low pass filter

```python
def gaussian_lpf(image,sigma):
    filter_ = np.zeros(image.shape)
    center = tuple(map(lambda x: x // 2, image.shape))
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            filter_[i, j] = gaussian_2d(i,j,center[0],center[1],sigma,sigma,1)
    
    ft_image = fourier_transform(image)
    ft_image = ft_image * filter_
    ift_image = scale(inverse_fourier_transform(ft_image))
    
    return ift_image
```

# defination of ideal high pass filter

```python
def ideal_hpf(image, radius):
    _filter = np.zeros(image.shape)
    center = tuple(map(lambda x: x // 2, image.shape))
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            if (i - center[0]) ** 2 + (j - center[1]) ** 2 >= radius ** 2:
                _filter[i, j] = 1 

    ft_image = fourier_transform(image)
    ft_image_filtered = ft_image * _filter
    ft_image = ft_image + ft_image_filtered
    ift_image = inverse_fourier_transform(ft_image)
    
    return ift_image
```

```python
def ideal_hpf2(image, radius):
    _filter = np.zeros(image.shape)
    center = tuple(map(lambda x: x // 2, image.shape))
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            if (i - center[0]) ** 2 + (j - center[1]) ** 2 >= radius ** 2:
                _filter[i, j] = 1 

    ft_image = fourier_transform(image)
    ft_image = ft_image * _filter
    ift_image = inverse_fourier_transform(ft_image)
    ift_image=image+ift_image
    
    return ift_image
```

```python
def ideal_hpf3(image, radius):
    blur_image=ideal_lpf(image,radius);
    image_edges=image-blur_image
    sharp_image=image+image_edges
    return sharp_image
```

# defination of gaussian high pass filter


```python
def gaussian_hpf(image, sigma):
    _filter = np.zeros(image.shape)
    center = tuple(map(lambda x: x // 2, image.shape))
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            _filter[i, j] = 1-gaussian_2d(i,j,center[0],center[1],sigma,sigma,1)

    ft_image = fourier_transform(image)

    ft_image_filtered = ft_image * _filter
    
    ft_image = ft_image + ft_image_filtered
    
    ift_image = inverse_fourier_transform(ft_image)
    
    return ift_image
```

# inverse of image

```python
def inverse(image):
    image_inverse=np.zeros(image.shape)
    n,m=image.shape[0],image.shape[1]
    for i in range(n):
        for j in range(m):
            image_inverse[i,j]=255-image[i,j]
    return image_inverse
            
```

# contrast streching function

```python
def contrast_streching(image,i1,i2,s):
    image_transformed = np.zeros(image.shape)
    n,m=image.shape[0],image.shape[1]
    for i in range(n):
        for j in range(m):
            if image[i,j] >= i1 and image[i,j] <= i2:
                image_transformed[i,j] = s*image[i,j]
            else :
                image_transformed[i,j] = image[i,j]
    scaled_image = scale(image_transformed)
    return scaled_image
```

# binary function 

```python
def binary(image,s):
    image_transformed = np.zeros(image.shape)
    n,m=image.shape[0],image.shape[1]
    for i in range(n):
        for j in range(m):
            if image[i,j] >= s:
                image_transformed[i,j] = 255
            else :
                image_transformed[i,j] = 0
    return image_transformed
```

```python
def bright(image,i1,i2):
    image_transformed = np.zeros(image.shape)
    n,m=image.shape[0],image.shape[1]
    for i in range(n):
        for j in range(m):
            if image[i,j] >= i1 and image[i,j] <= i2:
                image_transformed[i,j] = 255
            else :
                image_transformed[i,j] = image[i,j]
    return image_transformed
```

```python
def dim(image,i1,i2):
    image_transformed = np.zeros(image.shape)
    n,m=image.shape[0],image.shape[1]
    for i in range(n):
        for j in range(m):
            if image[i,j] >= i1 and image[i,j] <= i2:
                image_transformed[i,j] = 0
            else :
                image_transformed[i,j] = image[i,j]
    return image_transformed
```

# histogram equalization

```python
def histogram_equalization(image):
    n,m=image.shape
    image_eq=np.zeros(image.shape)
    histogram=np.zeros(256)
    for i in range(n):
        for j in range(m):
            histogram[image[i,j]]+=1
    for i in range(1,256,1):
        histogram[i]+=histogram[i-1]
    for i in range(0,256,1):
        histogram[i]/=histogram[255]
    for i in range(0,256,1):
        histogram[i]*=255
    for i in range(0,256,1):
        histogram[i]=round(histogram[i])
    for i in range(n):
        for j in range(m):
            image_eq[i,j]=histogram[image[i,j]]
    return image_eq
```
